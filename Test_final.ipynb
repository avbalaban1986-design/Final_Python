```python
# Прочитайте данные (переменную назовите 'df')
ab = pd.read_csv('data2.csv', sep=';')

# Вывести несколько первых строк таблицы данных
print(ab.head())
print(ab.columns)
```

```python
from scipy.stats import ttest_ind
experiments = sorted(ab['experiment_num'].unique())
```

```python
ab = ab.dropna(subset=['experiment_num'])
ab['experiment_num'] = ab['experiment_num'].astype(int)
```

```python
experiments = sorted(ab['experiment_num'].unique())

for exp in experiments:
    exp_data = ab[ab['experiment_num'] == exp]

    control = exp_data[exp_data['experiment_group'] == 'control']['revenue']
    test = exp_data[exp_data['experiment_group'] == 'test']['revenue']

    arpu_control = control.mean()
    arpu_test = test.mean()

    t_stat, p_value = ttest_ind(test, control, equal_var=False)

    print(f'\nЭксперимент {exp}')
    print(f'ARPU control: {arpu_control:.2f}')
    print(f'ARPU test: {arpu_test:.2f}')
    print(f'p-value: {p_value:.4f}')
```

```python
from statsmodels.stats.proportion import proportions_ztest

success = [1003, 1099]
nobs = [100047501, 100001055]

# Z-тест
z_stat, p_value = proportions_ztest(success, nobs)

print(z_stat)
print(p_value)

cr_A = success[0] / nobs[0]
cr_B = success[1] / nobs[1]

print(cr_A)
print(cr_B)

lift = (cr_B - cr_A) / cr_A * 100
print(lift)

if p_value < 0.05:
    print("Различия статистически значимы. Рекомендуется внедрить вариант B.")
else:
    print("Статистически значимых различий нет.")
```
